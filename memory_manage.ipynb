```python
import gc
class MyObj:
    def __init__(self,name):
        self.name=name
        print(f"{self.name} created")
    def __del__(self):
        print(f"object {self.name} destroyed")

obj1=MyObj("obj1")
obj2=MyObj("obj2")

# incase of circular reference
obj1.ref=obj2
obj2.ref=obj1
```

```python
del obj1
del obj2

gc.collect()  # to run garbage collector manually
```

```python
print(gc.garbage)
```

```python
# generators and memory management
# allows you to produce items one at a time and using memory efficiently by only keeping one item in memory at a time

def generator_numbers(n):
    for i in range(n):
        yield i

# using generators
for num in generator_numbers(10000):
    print(num)
    if num >= 10:
        break
```

```python
import tracemalloc

def create_list():
    return [i for i in range(10000)]

def main():
    tracemalloc.start()

    create_list()
    snapshot=tracemalloc.take_snapshot()
    top_stats=snapshot.statistics('lineno')

    print("[ Top 10 ] ")
    for stat in top_stats[:10]:
        print(stat)
```

```python
main()
```
